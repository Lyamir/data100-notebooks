Dataset used: List of European Countries by EU membership and Coastline


```python

import pandas as pd
import matplotlib.pyplot as plt
```

```python
countries = pd.read_csv('Countries.csv')
```

```python
fig = plt.figure()
caption1 ="This visualization shows the percentage of European countries"
caption2 ="which are member states of the European Union"
ax = countries['EU'].value_counts().plot(kind="pie",autopct='%1.0f%%', pctdistance=1.2,labels=None)

ax.set_title('Member States of the European Union', fontsize=14)
ax.set_ylabel('')
ax.set_xlabel('')
ax.legend(['Yes', 'No'], loc='upper right', fontsize=11)
fig.text(.5, .05, caption2, ha='center', fontsize=11)
fig.text(.5, .1, caption1, ha='center', fontsize=11)
plt.savefig('EU Countries.png')
```
